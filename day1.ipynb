```python
from dotenv import load_dotenv
import os
import requests
load_dotenv();
```

## Part 1

```python
headers={
    "Cookie": f"session={os.getenv('SESSION_COOKIE')}"
}
response = requests.get("https://adventofcode.com/2024/day/1/input", headers=headers)
response
```

```python
response.text
```

```python
import pandas as pd
from io import StringIO
data = pd.read_csv(StringIO(response.text), header=None, sep=r"\s+")
data.columns=['list1', 'list2']
data
```

```python
sorted1 = data['list1'].sort_values(ignore_index=True)
sorted2 = data['list2'].sort_values(ignore_index=True)
print(sorted1, sorted2)
```

```python
sorted_df = pd.concat([sorted1, sorted2], axis=1)
sorted_df.columns = ['list1', 'list2']
sorted_df['diff'] = sorted_df.apply(lambda x: abs(x['list2'] - x['list1']), axis=1)
sorted_df.astype({'list1':'int', 'list2':'int', 'diff':'int'})
sorted_df.dtypes
```

```python
sum(sorted_df['diff'])
```

### Part 2

```python
sorted_df
```

```python
# Count occurrences of each value in list2
list2_counts = sorted_df['list2'].value_counts()

# Map counts to list1 values
sorted_df['list2_count'] = sorted_df['list1'].map(list2_counts).fillna(0).astype(int)

print(sorted_df)
```

```python
sorted_df['similarity_score'] = sorted_df['list2_count'] * sorted_df['list1']
sorted_df
```

```python
sorted_df['similarity_score'].sum()
```
